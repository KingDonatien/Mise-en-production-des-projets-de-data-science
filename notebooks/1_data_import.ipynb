## loading data from s3 and saving it in the data folder

```python
import s3fs
import pandas as pd
import sys
sys.path.append('..')
from config import DATA_DIR

MY_BUCKET = "cmoreliere"
CHEMIN_FICHIER = "diffusion/data.csv"



fs = s3fs.S3FileSystem(client_kwargs={"endpoint_url": "https://minio.lab.sspcloud.fr"})

with fs.open(f"s3://{MY_BUCKET}/{CHEMIN_FICHIER}") as f:
    df_raw = pd.read_csv(f, index_col=0, parse_dates=True)

df_raw.index = pd.to_datetime(df_raw.index, utc=True)

df = df_raw["2024-01-01":]

# Enregistrer le dataframe en csv
df_raw.to_csv(f'{DATA_DIR}/df_raw.csv', index=True)
df_raw.head()
```
